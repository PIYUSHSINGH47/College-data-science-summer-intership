```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
np.random.seed(0)
n_samples=200
n_features=3
x=np.random.randn(n_samples, n_features)
n_components=1
pca=PCA(n_components=n_components)
```

```python
x_pca=pca.fit_transform(x)
x_orignal=pca.inverse_transform(x_pca)
plt.figure(figsize=(16,8))
plt.subplot(3,4,3)
plt.scatter(x[:,1],x[:,2],alpha=0.7)
plt.title('orignal data')
```

```python
plt.subplot(2,3,3)
plt.scatter(x_orignal[:,0],x_orignal[:,1],alpha=0.7)
plt.title('PCA-Reconstructed Data')
plt.tight_layout()
plt.show()
```
